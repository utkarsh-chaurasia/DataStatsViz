```python
import pandas as pd
import numpy as np

# Detect outliers using the IQR method
def detect_outliers(df, iqr_threshold=1.5):
  """Detects outliers using the IQR method.

  Args:
    df: A Pandas DataFrame.
    iqr_threshold: The IQR threshold.

  Returns:
    A Pandas Series of boolean values, indicating whether each row is an outlier.
  """

  q1 = df.quantile(0.25)
  q3 = df.quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - iqr * iqr_threshold
  upper_bound = q3 + iqr * iqr_threshold
  outliers = (df < lower_bound) | (df > upper_bound)
  return outliers

# Remove outliers from the DataFrame
df_outlier_free = df[~detect_outliers(df)]

# Save the outlier-free DataFrame to a new CSV file
df_outlier_free.to_csv('iris_outlier_free.csv', index=False)
```
